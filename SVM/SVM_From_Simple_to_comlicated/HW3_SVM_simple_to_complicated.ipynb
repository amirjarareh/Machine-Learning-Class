```python
import random
import numpy as np
import matplotlib.pyplot as plt
from  matplotlib.pyplot import get_cmap
from sklearn.svm import SVC
from mlxtend.plotting import plot_decision_regions
import pandas as pd
```

```python
import sklearn.datasets as skdt
from sklearn.inspection import DecisionBoundaryDisplay
import matplotlib.gridspec as gridspec
import itertools
```

# Blobs 2 Feature Simple

```python
X, y = skdt.make_blobs(n_samples=600, n_features=2,
                           centers=[[2, 2],[-2, -2]],
                           cluster_std=[2, 2], random_state=5)
gs = gridspec.GridSpec(2, 2)
```

## C changes

```python
clf1 = SVC(C = 1, kernel = "rbf",gamma='scale')
clf2 = SVC(C = 10, kernel = "rbf",gamma='scale')
clf3 = SVC(C = 50, kernel = "rbf", degree=3,gamma='scale')
clf4 = SVC(C = 100, kernel = "rbf",gamma='scale')
```

```python
fig = plt.figure(figsize=(10,8))

labels = ['rbf C = 1 kernel', 'rbf C = 10 kernel', 'rbf C = 50 kernel', 'rbf C = 100 kernel']
for clf, lab, grd in zip([clf1, clf2, clf3, clf4],
                         labels,
                         itertools.product([0, 1], repeat=2)):

    clf.fit(X, y)
    ax = plt.subplot(gs[grd[0], grd[1]])
    fig = plot_decision_regions(X=X, y=y, clf=clf,hide_spines = False, legend=1, filler_feature_values={2:2}, filler_feature_ranges={2:2})
    DecisionBoundaryDisplay.from_estimator(
    clf,
    X,
    plot_method="contour",
    colors="k",
    levels=[-1, 0, 1],
    alpha=0.5,
    linestyles=["--", "-", "--"],
    ax=ax,
    )
    plt.title(lab)

plt.show()
```

## kernel changes

```python
clf1 = SVC(kernel = "linear",gamma='scale')
clf2 = SVC(kernel = "rbf",gamma='scale')
clf3 = SVC(kernel = "poly", degree=3,gamma='scale')
clf4 = SVC(kernel = "sigmoid",gamma='scale')
```

```python
fig = plt.figure(figsize=(10,8))

labels = ['linear kernel', 'rbf kernel', 'poly with degree of 3 kernel', 'sigmoid kernel']
for clf, lab, grd in zip([clf1, clf2, clf3, clf4],
                         labels,
                         itertools.product([0, 1], repeat=2)):

    clf.fit(X, y)
    ax = plt.subplot(gs[grd[0], grd[1]])
    fig = plot_decision_regions(X=X, y=y, clf=clf,hide_spines = False, legend=1, filler_feature_values={2:2}, filler_feature_ranges={2:2})
    DecisionBoundaryDisplay.from_estimator(
    clf,
    X,
    plot_method="contour",
    colors="k",
    levels=[-1, 0, 1],
    alpha=0.5,
    linestyles=["--", "-", "--"],
    ax=ax,
    )
    plt.title(lab)

plt.show()
```

# Spiral

```python
from numpy import pi
N = 400
theta = np.sqrt(np.random.rand(N))*2*pi

r_a = 2*theta + pi
data_a = np.array([np.cos(theta)*r_a, np.sin(theta)*r_a]).T
x_a = data_a + np.random.randn(N,2)

r_b = -2*theta - pi
data_b = np.array([np.cos(theta)*r_b, np.sin(theta)*r_b]).T
x_b = data_b + np.random.randn(N,2)

res_a = np.append(x_a, np.zeros((N,1)), axis=1)
res_b = np.append(x_b, np.ones((N,1)), axis=1)


res = np.append(res_a, res_b, axis=0)
X = res[:,:-1]
y = res[:,-1].astype(np.integer)
gs = gridspec.GridSpec(2, 2)
```

## C changes

```python
clf1 = SVC(C = 1, kernel = "rbf",gamma='scale')
clf2 = SVC(C = 10, kernel = "rbf",gamma='scale')
clf3 = SVC(C = 50, kernel = "rbf", degree=3,gamma='scale')
clf4 = SVC(C = 100, kernel = "rbf",gamma='scale')
```
